```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(data=df)
plt.show()
```

```python
sns.scatterplot(data=df)
```

```python
q3 = df["height"].quantile(0.75)
q1 = df["height"].quantile(0.25)
iqr = q3 - q1
print(iqr)
```

```python
q1, q3 = df["height"].quantile([0.25, 0.75])
iqr = q3 - q1
lb, ub = q1 - 1.5*iqr, q3 + 1.5*iqr
outliers = df[(df["height"] < lb) | (df["height"] > ub)]

print(f"Lower bound: {lb}\nUpper bound: {ub}\nOutliers:\n{outliers}")
```

```python
import numpy as np
from scipy import stats

z = np.abs(stats.zscore(df["height"]))
z = pd.Series(z, name="height")   # make it a Series with name
print(z)
```

```python
df1 = df[z<3]
print(df1)
```

```python
q1=df.sepal_width.quantile(0.25)
q3=df.sepal_width.quantile(0.75)
iq=q3-q1
print(iq)
```

```python
rid=df[((df.sepal_width<(q1-1.5*iq))|(df.sepal_width>(q3+1.5*iq)))]
rid['sepal_width']
```

```python
delid=df[~((df.sepal_width<(q1-1.5*iq))|(df.sepal_width>(q3+1.5*iq)))]
delid
```

```python
sns.boxplot(x ='sepal_width', data=delid)
```
